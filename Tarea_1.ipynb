```python
def cuadro_magico(cuadro):
    # Verificamos las filas
    for fila in cuadro:
        if sum(fila) != Constante_magica:
            return False
    # Verificamos las columnas
    for col in range(n):
        if sum(cuadro[fila][col] for fila in range(n)) != Constante_magica:
            return False
    # Verificar diagonales
    if sum(cuadro[i][i] for i in range(n)) != Constante_magica:
        return False
    if sum(cuadro[i][n-i-1] for i in range(n)) != Constante_magica:
        return False
    return True
```
